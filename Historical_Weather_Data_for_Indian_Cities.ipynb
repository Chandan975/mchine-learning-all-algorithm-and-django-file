Detecting the outliers

```python
import seaborn as sns
sns.boxplot(df1['maxtempC'])
```

```python
print(where(df1['maxtempC']>10))
```

```python
Q1 = percentile(df1['maxtempC'], 25,
                   interpolation = 'midpoint')
 
Q3 = percentile(df1['maxtempC'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
IQR
```

```python
upper = where(df1['maxtempC'] >= (Q3+1.5*IQR))
# Lower bound
lower = where(df1['maxtempC'] <= (Q1-1.5*IQR))
 
''' Removing the Outliers '''
df1.drop(upper[0], inplace = True)
df1.drop(lower[0], inplace = True)
```
